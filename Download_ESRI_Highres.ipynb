## Option 1 : UI


```python
from useful_libs.basemap.wayback_viewer import WaybackViewer
from useful_libs.basemap.wayback_downloader import WaybackDownloader

# 1. Initialize the Map Viewer
# You can set the center to your specific Bengaluru coordinates
view = WaybackViewer(center=(12.9814, 77.7747), zoom=17)

# 2. Attach the Downloader logic
# This adds the Draw Tool and the Download UI to the map automatically
dl = WaybackDownloader(view)

# 3. Display the map
view.show()

### Now Draw a rectangle
### Click on Download images
```

## Option 2 :via Script (No Drawing)

```python
# Define your bounding box: (min_lat, min_lon, max_lat, max_lon)
my_bbox = (12.9801, 77.7734, 12.9827, 77.7759)

# Trigger the download manually at Zoom 18
dl.run_download(bbox=my_bbox, zoom=18, folder_name="whitefield_manual_export")
```
